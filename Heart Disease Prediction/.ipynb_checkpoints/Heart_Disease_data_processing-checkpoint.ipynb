### Heart disease prediction - Kaggle dataset
<br><br />
#### Data cleaning and exploration
Here we take a look at the kaggle dataset for [heart failure prediction](https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction?select=heart.csv), given a set of attributes like age, blood pressure, cholesterol, max heart rate etc. This part of the code takes a look to clean the data, address issues of missing or outlier values and create dummy attributes for non-numeric data so that it is suitable for modelling.<br><br />

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')    # reading original downloaded file
df.head(10)
```

```python
df.describe(include = 'all')
```

<br><br />
Looks like there is no missing data. This is not surprising given the kaggle usability rating of 10.0 for this dataset. But we do see that some values like _RestingBP_ and _Cholesterol_ have unrealistic extreme values. And we see that _ExerciseAngina_ has Y or N data for pain during exercise, which can be converted to 1 and 0 respectively.<br><br />

```python
df['ExerciseAngina'].replace({'N':0, 'Y':1}, inplace=True)
df.tail()
```

```python
sns.boxplot(x='HeartDisease', y='RestingBP', data=df)
```

```python
sns.boxplot(x='HeartDisease', y='Cholesterol', data=df)
```

<br><br />
_RestingBP_ (blood pressure) values of near 0 are clearly bad data, and so is 0 _Cholesterol_. So we replace these with mean values of the attributes. But first we must check how much data gets modified by this.<br><br />

```python
print('Problem values count for RestingBP: ', df[df["RestingBP"]<10]["RestingBP"].value_counts())
print('Problem values count for Cholesterol: ', df[df["Cholesterol"]==0]["Cholesterol"].value_counts())
```

<br><br />
Since there are 172 values (18% of total data inputs) to be considered for _Cholesterol_ correction, it is better to drop these rather than use average values because this could significantly affect the contribution of this attribute and introduce bias. We can safely replace the 1 problem values of _RestingBP_ with the average of the attribute.
<br><br />
Alternatively, since this dataset is not that large, we could create a parallel dataset with average values for 0 _Cholesterol_ values and use it to train parallel models and see if retaining those values added to the accuracy of the trained model. For this exercise, we will leave out this step.<br><br />

```python
# -- replacing RestingBP issues with avg values --
avg_rbp = df["RestingBP"].astype("float").mean()
df.replace(df[df["RestingBP"]<10]["RestingBP"], avg_rbp, inplace=True)

# -- dropping data with Cholesterol = 0 --
df.drop(df[df['Cholesterol']==0].index, inplace=True)
df.reset_index(drop=True, inplace=True)

df.describe(include = 'all')
```

<br><br />
Next we get dummy attributes for the attributes with 2 or more objects that cannot be reduced to 0 and 1.<br><br />

```python
df = pd.get_dummies(df, columns=['Sex', 'ChestPainType', 'RestingECG', 'ST_Slope'])
df.head()
```

```python
ax = sns.heatmap(df.corr(method='pearson'))
plt.show()

# -- correlation values for HeartDisease
cor = df.corr(method='pearson')['HeartDisease']
cor.sort_values()
```

<br><br />From the above correlation heatmap it is clear that some of the variables standout in correlating with _HeartDisease_, for eg. _ExerciseAngina_, _Oldpeak_, _ChestPainType_ASY_, _ST_Slope_Flat_ and _ST_Slope_Up_. We will finally check if these attributes play an important role in model contribution. Important to note that _Sex_M_ is weakly positively correlated to _HeartDisease_, and _Sex_F_ is weakly negatively. This could be a bit misleading since there is skewness in data in favor of _Sex_M_ (564 M, 182 F), so correlation values don't always lead to accurate insights at first glance.<br><br />

```python
df["Sex_M"].value_counts()
```

<br><br />
Finally we write the cleaned data with dummy attributes into a file to be used by Machine Learning models directly.<br><br />

```python
df.to_csv('heart_dumm_clean.csv')
```
